```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('nlogn_output.txt', 'r') as fin:
    nlogn_times = list(map(float, fin.read().splitlines()))
    
with open('lca_output.txt', 'r') as fin:
    lca_times = list(map(float, fin.read().splitlines()))
```

```python
step_size = 500
max_size = 650000
vec_sizes = list(range(step_size, max_size + 1, step_size))
```

```python
f, axs = plt.subplots(1, 2, figsize=(15, 7))

plt.subplot(1, 2, 1)
plt.plot(vec_sizes, nlogn_times)
plt.title('RMQ n*log(n)')
plt.xlabel('Array size')
plt.ylabel('Mean time')

plt.text(0, 0.25, f'Max size time divided\nby min size time:\n{(nlogn_times[-1] / nlogn_times[0]):.2f}')
plt.text(0, 0.2, f'Max size divided\nby min size:\n{(vec_sizes[-1] / vec_sizes[0]):.2f}')

plt.subplot(1, 2, 2)
plt.ylim(0, 1.5)
plt.plot(vec_sizes, lca_times)
plt.title('RMQ LCA')
plt.xlabel('Array size')
plt.ylabel('Mean time')

plt.text(0, 1, f'Max size time divided\nby min size time:\n{(lca_times[-1] / lca_times[0]):.2f}')
plt.text(0, 0.8, f'Max size divided\nby min size:\n{(vec_sizes[-1] / vec_sizes[0]):.2f}')
```
